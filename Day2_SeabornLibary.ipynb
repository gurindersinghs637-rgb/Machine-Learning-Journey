```python
df=sns.load_dataset("tips") # another way
df.head()
```

```python
sns.jointplot(x='tip',y="total_bill",data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y="total_bill",data=df,kind='reg')
```

```python
sns.pairplot(df,hue="sex")  # PAIRPLOT= USED FOR NUMERIC DATA
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
